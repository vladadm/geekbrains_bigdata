```python
import numpy as np
import pandas as pd
```

```python
from sklearn.datasets import load_boston
```

```python
boston_ds = load_boston()
```

```python
boston_ds.keys()
```

```python
boston_ds.feature_names
```

```python
X = pd.DataFrame(boston_ds.data, columns=boston_ds.feature_names)
```

```python
X
```

```python
y = pd.DataFrame(boston_ds.target, columns=['price'])
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=30, random_state=42 )
```

```python
X_train.info()
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lr = LinearRegression()
```

```python
lr.fit(X_train, y_train)
```

```python
price_predict = lr.predict(X_test)
price_predict
```

```python
from sklearn.metrics import r2_score
```

```python
linear_regress_r2_score = r2_score(y_test, price_predict)
linear_regress_r2_score
```
